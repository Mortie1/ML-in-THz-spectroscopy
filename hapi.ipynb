```python
from hapi import *
db_begin('data') # папка, куда будут скачаны линии поглощения
fetch('H2O', 1, 1, 3, 34) # скачивание линий: молекула, её номер в базе хитран, номер изотопа, диапазон длин волн
fetch('O3', 3, 1,  3, 34)
fetch('O2', 7, 1,  3, 34)
fetch('CH4', 6, 1, 3, 34)
fetch('SO2', 9, 1,  3, 34)
fetch('NH3', 11, 1,  3, 34)
fetch('NO', 8, 1,  3, 34)
fetch('H2S', 31, 1,  3, 34)
```

```python
def coef_spectra(compn, iso, comp, pressure): # функция для генерации множества смесей с разными параметрами с помощью цикла
    nu,coef = absorptionCoefficient_Voigt(((compn,iso),),comp, OmegaRange=(3, 34), OmegaStep=0.016, HITRAN_units=False, Environment={'p':pressure,'T':296}) # функция хитран для генерации спектра поглощения из скачаных линий, параметры - номер вещества в базе хитран и номер его изотопа, диапазон длин волн, шаг измерения, давление и температура
    return coef
```

```python
h2o_ = []
o3_ = []
o2_= []
ch4_ = []
so2_ = []
nh3_ = []
no_ = []
h2s_ = []



pres = np.linspace(0.01,2,50)

for item in pres:
    # for te in temp:
    h2o_.append(coef_spectra(1,1, 'H2O',item))
    o3_.append(coef_spectra(3,1, 'O3',item))
    o2_.append(coef_spectra(7,1, 'O2',item))
    ch4_.append(coef_spectra(6,1, 'CH4',item))
    so2_.append(coef_spectra(9,1, 'SO2',item))
    nh3_.append(coef_spectra(11,1, 'NH3',item))
    no_.append(coef_spectra(8,1, 'NO',item))
    h2s_.append(coef_spectra(31,1, 'H2S',item))
```

```python
X = np.concatenate([h2o_, o3_,o2_, ch4_, so2_, nh3_,no_, h2s_])
np.savetxt("gas_right2.txt", X)
```
